```python
import pandas as pd
```

```python
df = pd.read_csv(r"E-Scooter_Trips_20240215.csv")
df
```

```python
df=df.drop_duplicates()
df
```

```python
df['Start Community Area Name'] = df['Start Community Area Name'].str.strip()
df['End Community Area Name'] = df['End Community Area Name'].str.strip()
df['Vendor'] = df['Vendor'].str.strip()
df
```

```python
df = df.fillna('')
df
```

```python
for x in df.index:
  if df.loc[x, "End Community Area Name"]=='':
    df.drop(x,inplace=True)
df
```

```python
for x in df.index:
  if df.loc[x, "Start Community Area Name"]== '':
    df.drop(x,inplace=True)
df
```

```python
df['Start Centroid Location'] = df['Start Centroid Location'].str.replace('POINT \(', '').str.replace('\)', '').str.replace(' ', ', ')
df
```

```python
df['End Centroid Location'] = df['End Centroid Location'].str.replace('POINT \(', '').str.replace('\)', '').str.replace(' ', ', ')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
# Save the DataFrame to an Excel file
df.to_csv('cleaned_e-scooter_data.csv', index=False)
```

```python
import pandas as pd
import os

# List all uploaded CSV files
uploaded_files = [file for file in os.listdir('BSS data/') if file.endswith('.csv')]

# Initialize an empty list to store DataFrames
dfs = []

# Read each uploaded CSV file into a DataFrame and append it to the list
for file in uploaded_files:
    df = pd.read_csv('BSS data/' + file)
    dfs.append(df)

# Concatenate all DataFrames in the list along the rows
merged_df = pd.concat(dfs, ignore_index=True)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('bike_merged_data.csv', index=False)
```
