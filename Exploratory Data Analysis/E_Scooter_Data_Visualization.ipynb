**Total average distance travelled**

```python
# Calculate the average distance
# Convert distances from meters to kilometers
df['Trip Distance'] = df['Trip Distance'] / 1000

# Calculate the average distance in kilometers
Average_Distance_Kilometers = df['Trip Distance'].mean()
print(f"Average distance traveled: {Average_Distance_Kilometers} km")
```

**Most common start location**

```python
Most_Common_Start_Location = df['Start Community Area Name'].mode()[0]
print(f"Most common start location: {Most_Common_Start_Location}")
```

**Month with the highest number of rides**

```python
df['month'] = df['Start Time'].dt.month
# Count the occurrences of each month and find the month with the highest number of rides
Most_Preferred_Month = df['month'].value_counts().idxmax()
import calendar
Most_Preferred_Month_Name = calendar.month_name[Most_Preferred_Month]

print(f"The most preferred month for e-scooter riding is: {Most_Preferred_Month_Name}")
```

**Maximum distance travelled in a month**

```python
df.rename(columns={'month': 'Month'}, inplace=True)

# Sum the distances for each month
Monthly_Distance_Sum = df.groupby('Month')['Trip Distance'].sum()

# Find the month with the highest total distance
Most_Distance_Month = Monthly_Distance_Sum.idxmax()

# Translate the numeric month to a month name for readability (optional)
Most_Distance_Month_Name = calendar.month_name[Most_Distance_Month]

print(f"The month with the most distance travelled is: {Most_Distance_Month_Name} with a total distance of {Monthly_Distance_Sum.max()} kilometers.")
```

**Monthwise Usage Trends**

```python
df['Month'] = df['Start Time'].dt.to_period('M')

# Count trips by month
usage_trends = df.groupby('Month').size()

# Plot
usage_trends.plot(kind='bar')
plt.title('Usage Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Trips')
plt.xticks(rotation=45)
plt.show()
```

**Average Trip Duration and Distance**

**Feature Engineering: Create features that could affect demand, like time of day, day of the week, and holidays.**

```python
# Convert Start Time and End Time to datetime
df['Start Time'] = pd.to_datetime(df['Start Time'])
df['End Time'] = pd.to_datetime(df['End Time'])

# Extract features
df['start_hour'] = df['Start Time'].dt.hour
df['end_hour'] = df['End Time'].dt.hour
df['start_day_of_week'] = df['Start Time'].dt.dayofweek
df['end_day_of_week'] = df['End Time'].dt.dayofweek
```

**Aggregate Demand: Create a new DataFrame that aggregates the number of trips started per area and hour.**

```python
# Aggregate data to get the count of trips starting in each area per hour
demand_df = df.groupby(['Start Community Area Number', 'start_hour']).size().reset_index(name='trip_starts')
```

```python
pivot_table = demand_df.pivot(index='Start Community Area Number', columns='start_hour', values='trip_starts')

# Use a heatmap to visualize the pivot table
plt.figure(figsize=(15, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=.5, annot=False)  # Set 'annot' to True if you want to see numbers
plt.title('Heatmap of Scooter Demand per Area and Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Community Area Number')
plt.show()
```
