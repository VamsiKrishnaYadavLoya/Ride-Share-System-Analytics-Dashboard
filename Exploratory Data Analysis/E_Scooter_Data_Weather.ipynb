Number of Rides vs. weather

```python
import matplotlib.pyplot as plt
import numpy as np

# Grouping the e-scooter data by 'weather' and counting the number of rides ('Trip ID') for each weather
rides_per_weather = df.groupby('weather')['ride_id'].count().reset_index()

# Plotting the graph
plt.figure(figsize=(5, 6))
plt.bar(rides_per_weather['weather'], rides_per_weather['ride_id'], color='blue')
plt.title('Number of Rides vs. weather')
plt.xlabel('weather (Average Temperature °F)')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.show()
```

Ride Duration vs. weather

```python
# Calculate the duration in minutes for each ride
df['duration'] = (pd.to_datetime(df['ended_at']) - pd.to_datetime(df['started_at'])).dt.total_seconds() / 60

# Group by 'weather' and calculate the average duration for each weather
average_duration_per_weather = df.groupby('weather')['duration'].mean().reset_index()

# Plotting the graph
plt.figure(figsize=(12, 5))
plt.plot(average_duration_per_weather['weather'], average_duration_per_weather['duration'], marker='o', linestyle='-', color='orange')
plt.title('Average Ride Duration vs. weather')
plt.xlabel('weather (Average Temperature °F)')
plt.ylabel('Average Duration (minutes)')
plt.xticks(rotation=45)
plt.show()
```

Number of Rides by Vendor Type vs. weather

```python
ride_counts_per_weather_vendor = df.groupby(['weather', 'Vendor'])['Trip ID'].count().unstack().fillna(0)

# Plotting the graph
ride_counts_per_weather_vendor.plot(kind='bar', figsize=(14, 7), width=0.8)
plt.title('Number of Rides by Vendor Type vs. weather')
plt.xlabel('weather (Average Temperature °F)')
plt.ylabel('Number of Rides')
plt.xticks(rotation=45)
plt.legend(title='Vendor Type')
plt.show()
```
